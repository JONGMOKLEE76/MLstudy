### 아래와 같이 구체적으로 지정이 가능함
 - lw : linewidth
 - ls : linestyle
 - ms : markersize
 - mec : markeredgecolor
 - mew : markeredgewidth
 - mfc : markerfacecolor

```python
plt.title('plotEx')
plt.plot([10, 20, 30, 40], [1, 3, 8, 9], c='b', lw=5, ls='--', marker='o', ms=10, mec='r', mew='3', mfc='g' )
plt.show()
```

```python
plt.title('plotEx')
plt.plot([10, 20, 30, 40], [1, 3, 8, 9], c='b', lw=5, ls='--', marker='o', ms=10, mec='r', mew='3', mfc='g' )
plt.xlim(0, 50) # x 값의 범위를 지정
plt.ylim(-5, 25) # y 값의 범위를 지정
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256) # x 값을 256개로 설정
y = np.cos(x)
plt.plot(x, y)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], # x값의 범위 중 원하는 지점만 선택할 떄
          ['-pi', '-pi/2', '0', 'pi/2', 'pi']) # 해당하는 x값의 명칭으로 나타낼 수도 있음
plt.yticks([-1, 0, 1], ['low', '0', 'high']) # 해당하는 y값의 명칭으로 나타낼 수도 있음
plt.show()
```

```python
# 한 그래프에 2가지 이상 그릴 때, 라벨 붙이고, 범례 표시
y1, y2 = np.cos(x), np.sin(x)
plt.plot(x, y1, ls='--', label='cosine')
plt.plot(x, y2, ls='--', label='sine')
plt.legend(loc='best')
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

### histogram 그리기

```python
x1 = 100 + 15 * np.random.randn(1000)
x2 = 140 + 15 * np.random.randn(1000)

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(1, 1, 1) 
ax1.hist(x1, bins=50, color='darkgreen') # bins :구간 몇개를 나눌 것인지를 나타냄
ax1.hist(x2, bins=50, color='orange', alpha=0.5)
ax1.set_xlabel('bins')
ax1.set_ylabel('distributions')
plt.show()
```

### Boxplot 그리기

### boxplot 의 최대, 최소값
 - 최대, 최소값의 경우, 아래와 같이 IQR 값의 1.5배만큼 더하거나 뺐을 때의 값이 최대/최소가 가질수 있는 최대 경계값이 되는 것임
 - 즉, 데이타의 최대/최소값은 최대 1.5*IQR 값이 되는 것이고 데이타의 실제 값에 따라 최대/최소는 이 범위 안에 있게 됨

```python
np.random.seed(777)
normal = np.random.normal(loc=0.0, scale=1.0, size=500)
lognormal = np.random.lognormal(mean=0.0, sigma=1.0, size=500)

box_plot_data = [normal, lognormal]

fig = plt.figure(figsize=(7, 5))
ax1 = fig.add_subplot(111)
ax1.boxplot(box_plot_data, vert=True, showmeans=True, labels=['Normal', 'Lognormal']) # vert 값이 false이면 horizontal 형태, showmeans 는 평균값 표시, label 은 라벨명을 표시
ax1.set_xlabel('distribution')
ax1.set_ylabel('value')
plt.show()
```

### 데이터프레임을 그대로 plot 하는 방법

```python
frame = pd.DataFrame(np.random.rand(5, 3),
                    index=['c1', 'c2', 'c3', 'c4', 'c5'],
                    columns=['metric1', 'metric2', 'metric3'])
frame
```

```python
fig, axes = plt.subplots(1, 2)
ax1, ax2 = axes.ravel() # ravel 함수는 다차원을 1차원으로 풀어줌(unpacking의 개념)
frame.plot(kind='bar', ax=ax1, alpha=0.75, title='bar plot')
ax1.set_xlabel('metric')
ax1.set_ylabel('value')
frame.plot(kind='box', ax=ax2, title='box plot')
ax2.set_xlabel('metric')
ax2.set_ylabel('value')
plt.tight_layout()
plt.show()
```
