### 선형 회귀

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100, 1) # rand 함수는 uniform distribution 을 보이는 0~1 범위에서 선택
x = x * 4 - 2
y = 3 * x - 2
y += np.random.randn(100, 1) # randn 함수는 Normal(Gaussian) distribution 을 보이는 -1~1 범위에서 선택

plt.scatter(x, y, marker='+')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() # 모델 생성
model.fit(x, y) # 학습
pred_y = model.predict(x) # 모델의 예측값
x_new = np.linspace(-2, 2, 2).reshape(2, 1)
pred_y_new = model.predict(x_new)
plt.scatter(x, y, marker='+') 
plt.plot(x_new, pred_y_new, color='r')
plt.show()
```

```python
# 모델의 계수, 절편 구하기
print('계수 : ', model.coef_)
print('절편: ', model.intercept_)
```

## 비선형 모델의 경우, 다항회귀(polynomial feature 사용)

```python
x = 6 * np.random.rand(100, 1) - 3
y = 0.5 * x**2 + x + 2 + np.random.randn(100, 1)

plt.plot(x, y, 'b.')
plt.axis([-3, 3, 0, 10])
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False) # Bias를 포함하지 않을 때
x_poly = poly_features.fit_transform(x)
```

```python
x[0], x_poly[0] # x 데이터가 polynomial feature를 통해 이와 같이 변함
```

```python
model2 = LinearRegression()
model2.fit(x_poly, y)

x_new = np.linspace(-3, 3, 100).reshape(100, 1)
x_new_poly = poly_features.transform(x_new)
y_new_pred = model2.predict(x_new_poly)
plt.plot(x, y, 'b.', label='data')
plt.plot(x_new, y_new_pred, 'r-', linewidth=2, label='prediction')
plt.legend(loc='best')
plt.axis([-3, 3, 0, 10])
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, x, y): # train 데이타의 사이즈를 점점 크게할 때, validation 에러가 점점 작아지는 것을 볼 수 있음
    x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
    train_errors, val_errors = [], []
    
    for m in range(1, len(x_train)):
        model.fit(x_train[:m], y_train[:m])
        y_train_predict = model.predict(x_train[:m])
        y_val_predict = model.predict(x_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
    
    plt.plot(np.sqrt(train_errors), 'r-+', linewidth=2, label='train')
    plt.plot(np.sqrt(val_errors), 'b-', linewidth=2, label='val')
    plt.legend(loc='best', fontsize=14)
    plt.xlabel('train set size', fontsize=14)
    plt.ylabel('MSE', fontsize=14)
    
model3 = LinearRegression()
plot_learning_curves(model3, x, y)
```
